```python
apple_img = cv2.imread(apple_image_path, cv2.IMREAD_COLOR)

plt.imshow(apple_img)
```

```python
apple_img_rgb = cv2.cvtColor(apple_img, cv2.COLOR_BGR2RGB)
plt.imshow(apple_img_rgb)
```

```python
r, g, b = cv2.split(apple_img_rgb)

plt.figure(figsize=[20, 5])

plt.subplot(141); plt.imshow(r, cmap='gray'); plt.title("red channel")
plt.subplot(142); plt.imshow(g, cmap='gray'); plt.title("green channel")
plt.subplot(143); plt.imshow(b, cmap='gray'); plt.title("blue channel")

apple_img_merged = cv2.merge((r, g, b))


plt.subplot(144); plt.imshow(apple_img_merged, cmap='gray'); plt.title("merged img")
print(plt.colormaps)
```

```python
img_apple_hsv = cv2.cvtColor(apple_img_rgb, cv2.COLOR_RGB2HSV)

img_apple_rgb = cv2.cvtColor(img_apple_hsv, cv2.COLOR_HSV2RGB)

h, s, v = cv2.split(img_apple_hsv)

plt.figure(figsize=[20, 5])

plt.subplot(141); plt.imshow(h, cmap='gray'); plt.title("hue channel")
plt.subplot(142); plt.imshow(s, cmap='gray'); plt.title("saturation channel")
plt.subplot(143); plt.imshow(v, cmap='gray'); plt.title("value channel")

img_apple_hsvmerge = cv2.merge((h, s, v))

plt.subplot(144); plt.imshow(img_apple_rgb, cmap='gray'); plt.title("original img")
```

```python
h_new = (h + 90) % 180

merged_new = cv2.merge((h_new, s, v))
merged_rgb = cv2.cvtColor(merged_new, cv2.COLOR_HSV2RGB)

plt.imshow(merged_rgb)
```

```python
apple_bgr = cv2.cvtColor(merged_rgb, cv2.COLOR_RGB2BGR)

out_path = os.path.join('output', 'apple_blue.png')
cv2.imwrite(out_path, apple_bgr)

IpyImage(out_path)
```
