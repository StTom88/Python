```python
import os
import cv2
import matplotlib.pyplot as plt

res_folder_path = os.path.join('resources')
img_path = os.path.join(res_folder_path, "dog.jpg")

img_bgr = cv2.imread(img_path)

# img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
# plt.imshow(img_rgb)

plt.imshow(img_bgr[:, :, ::-1]) # slicovani z BGR do RGB


```

```python
img_with_line = img_bgr.copy()

pt1 = (850, 100)
pt2 = (950, 600)
line_color = (255, 0, 0)

cv2.line(img_with_line, pt1, pt2, line_color[::-1], 4, cv2.LINE_AA)

plt.imshow(img_with_line[:, :, ::-1])
```

```python
img_with_circle = img_bgr.copy()

center = (360, 380)
circle_color = (255, 0, 0)

cv2.circle(img_with_circle, center, 100, circle_color[::-1], -1, cv2.LINE_AA)

plt.imshow(img_with_circle[:,:,::-1])
```

```python
img_with_rectangle = img_bgr.copy()

pt1_rect = (240, 100)
pt2_rect = (720, 580)
rectangle_color = (0, 255, 0)

cv2.rectangle(img_with_rectangle, pt1_rect, pt2_rect, rectangle_color[::-1], 4, cv2.LINE_AA)

plt.imshow(img_with_rectangle[:,:,::-1])
```

```python
img_with_text = img_bgr.copy()

text = "Whoof"
text_color = (255, 0, 0)
font = cv2.FONT_ITALIC
origin = (350, 670)

cv2.putText(img_with_text, text, origin, font, 3.5, text_color[::-1], 7, cv2.LINE_AA)

plt.imshow(img_with_text[:,:,::-1])
```

```python
cropped_img = img_bgr[200:800, :600, :]


resized_image_by_factor = cv2.resize(cropped_img, None, fx=2, fy=2)
resized_image_by_size = cv2.resize(cropped_img, (200, 200))

plt.figure(figsize=(15, 6))
plt.subplot(131); plt.imshow(cropped_img[:,:,::-1]), plt.title('Original')
plt.subplot(132); plt.imshow(resized_image_by_factor[:,:,::-1]), plt.title('Resized by factor')
plt.subplot(133); plt.imshow(resized_image_by_size[:,:,::-1]), plt.title('Resized by size')
print(cropped_img.shape)
```

```python
desired_width = 100

aspect_ratio = desired_width / cropped_img.shape[1]
print(aspect_ratio)

computed_height = int(cropped_img.shape[0] * aspect_ratio)
computed_height

dim = (desired_width, computed_height)

resized_img = cv2.resize(cropped_img, dim, interpolation=cv2.INTER_AREA)

plt.imshow(resized_img[:,:,::-1])
```

```python
img_flipped = img_bgr.copy()

img_flipped_hor = cv2.flip(img_flipped, 1)
img_flipped_ver = cv2.flip(img_flipped, 0)
img_flipped_both = cv2.flip(img_flipped, -1)

plt.figure(figsize=(20, 10))
plt.subplot(141); plt.imshow(img_flipped[:,:,::-1]), plt.title('Original')
plt.subplot(142); plt.imshow(img_flipped_hor[:,:,::-1]), plt.title('Image flipped horizontaly')
plt.subplot(143); plt.imshow(img_flipped_ver[:,:,::-1]), plt.title('RImage flipped verticaly')
plt.subplot(144); plt.imshow(img_flipped_both[:,:,::-1]), plt.title('Image flipped both')
```
